In the first part, I choose Mexico as my target. 
My data is collected from The World Bank. 
Regarding the currency, I set all currency as US dollar.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
gdp_capita = pd.read_csv('/content/drive/MyDrive/Macro/HW1/GDP per capita.csv', header=2, index_col=0)
gdp = pd.read_csv('/content/drive/MyDrive/Macro/HW1/GDP.csv', header=2, index_col=0)
capital = pd.read_csv('/content/drive/MyDrive/Macro/HW1/Gross capital formation.csv', header=2, index_col=0)

population = pd.read_csv('/content/drive/MyDrive/Macro/HW1/Population.csv', header=2, index_col=0)
fertility_rate = pd.read_csv('/content/drive/MyDrive/Macro/HW1/Fertility rate.csv', header=2, index_col=0)
life_ex = pd.read_csv('/content/drive/MyDrive/Macro/HW1/Life expectancy at birth.csv', header=2, index_col=0)
old_dep = pd.read_csv('/content/drive/MyDrive/Macro/HW1/Age dependency ratio.csv', header=2, index_col=0)
```

```python
# DElete unuseful columns
gdp_capita = gdp_capita.drop(columns={'Country Code', 'Indicator Name', 'Indicator Code', 'Unnamed: 65'})
print(gdp_capita.columns)

# Focus on two countries we care
mex = gdp_capita.loc['Mexico',:]
usa = gdp_capita.loc['United States']
```

```python
plt.style.use('seaborn')

tick_spacing = 1
tick_spacing = 5

_ = sns.lineplot(x=mex.index, y=mex)
_ = sns.lineplot(x=mex.index, y=usa)
_.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
plt.show()
```

(a) GDP questions

```python
# (i) GDP per capita (MEX) / GDP per capita(USA)
k = mex/usa *100

tick_spacing = 1
tick_spacing = 5

_ = sns.lineplot(x=k.index, y=k)
_.set_xlabel('Year')
_.set_ylabel('Ratio in US dollar(%)')


_.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))

plt.figure(figsize=(15,10))
plt.show()
```

```python
# Capital
capital = capital.drop(columns=['Country Code', 'Indicator Name', 'Indicator Code', 'Unnamed: 65'])
capital = capital.loc['Mexico', :]

# GDP
gdp = gdp.drop(columns=['Country Code', 'Indicator Name', 'Indicator Code', 'Unnamed: 65'])
gdp = gdp.loc['Mexico', :]
```

```python
# plot capital / gdp
k = capital / gdp *100

tick_spacing = 1
tick_spacing = 5

_ = sns.lineplot(x=k.index, y=k)
_.set_xlabel('Year')
_.set_ylabel('capital / gdp (%)')


_.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))


plt.show()
```

(b) Demographic structure

```python
print(population.columns)
```

```python
# (i) populatioin
k = population.drop(columns=['Country Code', 'Indicator Name', 'Indicator Code', 'Unnamed: 65'])
k = k.loc['Mexico', :]
k = k/1000

tick_spacing = 1
tick_spacing = 5

_ = sns.lineplot(x=k.index, y=k)
_.set_xlabel('Year')
_.set_ylabel('Population (per thousand)')

_.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))


plt.show()
```

```python
# (ii) fertility rate
k = fertility_rate.drop(columns=['Country Code', 'Indicator Name', 'Indicator Code', 'Unnamed: 65'])
k = k.loc['Mexico', :]

tick_spacing = 1
tick_spacing = 5

_ = sns.lineplot(x=k.index, y=k)
_.set_xlabel('Year')
_.set_ylabel('Fertility rate (births per woman)')

_.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))


plt.show()
```

```python
# (iii) Life expectancy
k = life_ex.drop(columns=['Country Code', 'Indicator Name', 'Indicator Code', 'Unnamed: 65'])
k = k.loc['Mexico', :]

tick_spacing = 1
tick_spacing = 5

_ = sns.lineplot(x=k.index, y=k)
_.set_xlabel('Year')
_.set_ylabel('Life expectancy (Number of years)')

_.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))


plt.show()
```

```python
# (iv) old-age dependency ratio
k = old_dep.drop(columns=['Country Code', 'Indicator Name', 'Indicator Code', 'Unnamed: 65'])
k = k.loc['Mexico', :]

tick_spacing = 1
tick_spacing = 5

_ = sns.lineplot(x=k.index, y=k)
_.set_xlabel('Year')
_.set_ylabel('Old-age dependency ratio (%)')

_.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))


plt.show()
```
